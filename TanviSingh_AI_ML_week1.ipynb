```python
# Python Fundamentals for AI/ML â€“ Week 1

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Arithmetic Operations
print(f"Addition: {num1 + num2}")
print(f"Subtraction: {num1 - num2}")
print(f"Multiplication: {num1 * num2}")
print(f"Division: {num1 / num2}")
print(f"Modulus: {num1 % num2}")
print(f"Power: {num1 ** num2}")

# Comparison & Logical Operators
print(f"{num1} > {num2}: {num1 > num2}")
print(f"{num1} < {num2}: {num1 < num2}")
print(f"{num1} == {num2}: {num1 == num2}")
print(f"{num1} != {num2}: {num1 != num2}")
print(f"Both positive (and): {num1 > 0 and num2 > 0}")
print(f"Any positive (or): {num1 > 0 or num2 > 0}")
print(f"Not positive: {not(num1 > 0)}")
```

```python
# Task -2 Lists And Arrays
# This task focuses on working with lists by adding, removing, and sorting elements.
# It also introduces NumPy arrays and basic statistical operations like mean, median, and standard deviation, which are commonly used in data analysis.

import random
import numpy as np

# Creating list pf random integers
numbers = [random.randint(1, 100) for _ in range(10)]
print("\nOriginal List:", numbers)

# Adding & Removing elements
numbers.append(55)
numbers.pop(0)
print("Updated List:", numbers)

# Find Min and Max
print("Max value:", max(numbers))
print("Min value:", min(numbers))

# Sorting the List
numbers.sort()
print("Sorted List:", numbers)

num_array = np.array(numbers)

print("\nMean:", np.mean(num_array))
print("Median:", np.median(num_array))
print("Standard Deviation:", np.std(num_array))
```

```python
# Task-3 Dictionaries and Sets
# This task explains how dictionaries store data in key-value pairs and how sets manage unique values.
# It also shows how conditions can be used to assign grades and how set operations like union and intersection work.

student = {}
student["name"] = "Tanvi"
student["course"] = "AI/ML"
student["marks"] = 80

if student["marks"] >= 75:
    student["grade"] = "A"
else:
    student["grade"] = "B"

for key in student:
    print(key, ":", student[key])

set1 = {"Python", "TensorFlow", "PyTorch"}
set2 = {"PyTorch", "OpenCV", "Keras"}

print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
```
